```python
import numpy as np
import matplotlib.pyplot as plt

# Mean and Standard Deviation
mean = 500
std_dev = 100

# Generate 1000 random numbers
data = np.random.normal(mean, std_dev, 1000)

# Plot Histogram
plt.hist(data, bins=30, edgecolor='black')
plt.title("Normal Distribution (Mean=500, Std=100)")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.grid(True)

plt.show()
```

```python
# Given Values
x = 9500      # Specific day's steps
mean = 8000
std_dev = 1200

# Z-Score Formula
z_score = (x - mean) / std_dev

print("Z-Score =", z_score)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# X-axis values
x = np.linspace(-4, 4, 1000)

# Standard Normal Distribution
y = norm.pdf(x, 0, 1)

# Plot Curve
plt.plot(x, y, color='blue')

# Shade Area
x_fill = np.linspace(-1, 1, 500)
plt.fill_between(x_fill, norm.pdf(x_fill, 0, 1), color='lightblue')

plt.title("Standard Normal Distribution")
plt.xlabel("Z-Score")
plt.ylabel("Probability Density")
plt.grid(True)

plt.show()
```

```python
from scipy.stats import norm

mean = 30
std_dev = 5


probability = norm.cdf(25, mean, std_dev)

print("Probability =", probability)
```

The normal distribution is important because many real-world data values naturally follow a bell-shaped pattern. Most values are close to the average, while very high and very low values are less common. In apps like Zomato, delivery times are usually around the average time, with only a few deliveries being much faster or slower. In sales analysis, most daily sales are near the average, while extremely high or low sales happen less often. User ratings also tend to cluster around common values. The normal distribution helps businesses understand their data, detect unusual values, calculate probabilities, make predictions, and improve decision-making. It is widely used in statistics, machine learning, finance, healthcare, and business analytics.
